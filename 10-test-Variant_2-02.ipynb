3. Имеется 2 банка. Первый банк ежемесячно начисляет 1.1% от первоначально вложенной суммы (простые проценты). Второй банк ежемесячно начисляет 1% от накопленной к этому моменту суммы (сложные проценты). В каждый из банков в месяц 0 (начальный момент) положили 100 000 руб. Напишите программу, которая выводит на экран в виде таблицы, накопленные в этих банках суммы для 24 месяцев. Определите в программе и выведите на экран в какой банк выгоднее вложить деньги на 1 год и на 2 года.

````python
# ваш код здесь
N = 24
dep = 100_000
vk1 = 0
vk2_summa = 100_000
for i in range(N):
    vk1+=dep*0.011
    print('|{:6}  |  {:6}  |'.format(i+1, dep*0.011))
print()
print('Сумма по первому вкаладу за 2 года', vk1+dep)
print()
nach = 100_000
for i in range(N):
    vk2_summa*=1.01
    print('|{:6}  |  {:6.2f}  |'.format(i+1, (vk2_summa - nach)))
    nach = vk2_summa
print()
print('Сумма по второму вкаладу за 2 года',round(vk2_summa,1))
print()
print()
N = 12
vk1 = 0
vk2_summa = 100_000
for i in range(N):
    vk1+=dep*0.011
    print('|{:6}  |  {:6}  |'.format(i+1, dep*0.011))
print()
print('Сумма по первому вкаладу за 1 год', vk1+dep)
print()
nach = 100_000
for i in range(N):
    vk2_summa*=1.01
    print('|{:6}  |  {:6.2f}  |'.format(i+1, (vk2_summa - nach)))
    nach = vk2_summa
print()
print('Сумма по второму вкаладу за 1 год',round(vk2_summa,1))
print()

````
